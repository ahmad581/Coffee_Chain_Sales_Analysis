```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
coffyChain_dataset = pd.read_csv('/kaggle/input/coffee-chain-sales-dataset/Coffee_Chain_Sales .csv')
dataset = coffyChain_dataset.copy()
dataset
```

```python
dataset.info()
```

```python
dataset.describe()
```

Sales Performance Analysis

1- total sales fot the entire dataset

```python
total_sales_for_the_entire_dataset = dataset['Sales'].sum()
total_sales_for_the_entire_dataset
```

2- How do sales vary by state

```python
state_sales_dataset = dataset.copy()
state_sales_dataset = state_sales_dataset.groupby(['State'])['Sales'].sum().reset_index()
state_sales_dataset
```

```python
plt.bar(state_sales_dataset['State'], state_sales_dataset['Sales'])
plt.xlabel('State')
plt.ylabel('Sales')
plt.title('Sales per State')
plt.xticks(rotation='vertical')
plt.show()
```

3- How do sales vary by market

```python
market_sales_dataset = dataset.copy()
market_sales_dataset = market_sales_dataset.groupby(['Market'])['Sales'].sum().reset_index()
market_sales_dataset
```

```python
plt.bar(market_sales_dataset['Market'], market_sales_dataset['Sales'])
plt.xlabel('Market')
plt.ylabel('Sales')
plt.title('Sales per Market')
plt.xticks(rotation='vertical')
plt.show()
```

4- How do sales vary by product line

```python
product_line_sales_dataset = dataset.copy()
product_line_sales_dataset = product_line_sales_dataset.groupby(['Product_line'])['Sales'].sum().reset_index()
product_line_sales_dataset
```

```python
plt.bar(product_line_sales_dataset['Product_line'], product_line_sales_dataset['Sales'])
plt.xlabel('Product_line')
plt.ylabel('Sales')
plt.title('Sales per Product_line')
plt.xticks(rotation='vertical')
plt.show()
```

5- How do sales vary by Date

```python
date_sales_dataset = dataset.copy()
date_sales_dataset['Date'] = pd.to_datetime(date_sales_dataset['Date'])
date_sales_dataset = date_sales_dataset.groupby(['Date'])['Sales'].sum().reset_index()
date_sales_dataset['Year'] = pd.DatetimeIndex(date_sales_dataset['Date']).year
date_sales_dataset['Month'] = pd.DatetimeIndex(date_sales_dataset['Date']).month
date_sales_dataset = date_sales_dataset.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
date_sales_dataset_splitted = [date_sales_dataset[date_sales_dataset['Year'] == x] for x in date_sales_dataset['Year'].unique()]
date_sales_dataset_splitted
```

```python
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(2, 2, figsize=(10, 10))

for i, ax in enumerate([ax0, ax1, ax2, ax3]):
    # ax.bar(date_sales_dataset_splitted[i].Month, date_sales_dataset_splitted[i].Sales)
    x = np.arange(len(date_sales_dataset_splitted[i].Month))  # Create an array from 0 to the number of months
    ax.bar(x, date_sales_dataset_splitted[i]['Sales'])
    ax.set_title(str(2012 + i))  # Assuming the titles correspond to the years
    ax.set_xlabel('Month')
    ax.set_ylabel('Sales')

    # Set the x-tick positions and labels
    ax.set_xticks(range(len(date_sales_dataset_splitted[i].Month)))
    ax.set_xticklabels(date_sales_dataset_splitted[i].Month)
plt.tight_layout()
plt.show()
```

Profit Analysis

1- the total profit for the entire dataset

```python
total_profit_for_the_entire_dataset = dataset['Profit'].sum()
total_profit_for_the_entire_dataset
```

2- profit compare to the target profit for each state

```python
profit_to_target_profit_dataset = pd.DataFrame()
profit_to_target_profit_dataset['State'] = dataset['State']
profit_to_target_profit_dataset['Profit'] = dataset['Profit']
profit_to_target_profit_dataset['Target_profit'] = dataset['Target_profit']
profit_to_target_profit_dataset
```

```python
groupby_profit = profit_to_target_profit_dataset.groupby(['State'])[['Profit', 'Target_profit']].sum().reset_index()
groupby_profit
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

bar1 = ax.barh(np.arange(len(groupby_profit)) - 0.35/2, groupby_profit['Profit'], 0.35, label='Profit')
bar2 = ax.barh(np.arange(len(groupby_profit)) + 0.35/2, groupby_profit['Target_profit'], 0.35, label='Target Profit')

ax.set_ylabel('State')
ax.set_xlabel('Profit')
ax.set_title('Profit vs. Target Profit by State')
ax.set_yticks(np.arange(len(groupby_profit)))
ax.set_yticklabels(groupby_profit['State'])
ax.invert_yaxis()
ax.legend()

for bar in [bar1, bar2]:
    for value in bar:
        width = value.get_width()
        ax.annotate('{}'.format(width),
                    xy=(width, value.get_y() + value.get_height() / 2),
                    xytext=(3, 0),  # 3 points horizontal offset
                    textcoords="offset points",
                    ha='left', va='center')
plt.tight_layout()
plt.show()
```

3- products with consistently low profits

```python
products_with_low_profit = pd.DataFrame()
products_with_low_profit['State'] = dataset['State']
products_with_low_profit['Product'] = dataset['Product']
products_with_low_profit['Sales'] = dataset['Sales']
products_with_low_profit['Profit'] = dataset['Profit']
products_with_low_profit = products_with_low_profit.groupby(['Product', 'State'])[['Sales', 'Profit']].sum().reset_index()
products_with_low_profit = [products_with_low_profit[products_with_low_profit['State'] == x] for x in products_with_low_profit['State'].unique()]
products_with_low_profit
```

```python
fig, ((ax0, ax1, ax2, ax3), (ax4, ax5, ax6, ax7), (ax8, ax9, ax10, ax11), (ax12, ax13, ax14, ax15), (ax16, ax17, ax18, ax19)) = plt.subplots(5, 4, figsize=(10, 10))

for i, ax in enumerate([ax0, ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9, ax10, ax11, ax12, ax13, ax14, ax15, ax16, ax17, ax18, ax19]):
    fig1, ax = plt.subplots(figsize=(10, 10))

    bar1 = ax.barh(np.arange(len(products_with_low_profit[i]['Profit'])) - 0.35/2, products_with_low_profit[i]['Profit'], 0.35, label='Profit')
    bar2 = ax.barh(np.arange(len(products_with_low_profit[i]['Sales'])) + 0.35/2, products_with_low_profit[i]['Sales'], 0.35, label='Sales')

    ax.set_ylabel('Products')
    ax.set_xlabel('Profit')
    ax.set_title(products_with_low_profit[i]['State'].iloc[0])
    ax.set_yticks(np.arange(len(products_with_low_profit[i]['Product'])))
    ax.set_yticklabels(products_with_low_profit[i]['Product'])
    ax.invert_yaxis()
    ax.legend()

    for bar in [bar1, bar2]:
        for value in bar:
            width = value.get_width()
            ax.annotate('{}'.format(width),
                        xy=(width, value.get_y() + value.get_height() / 2),
                        xytext=(3, 0),  # 3 points horizontal offset
                        textcoords="offset points",
                        ha='left', va='center')

plt.tight_layout()
plt.show()
```

Cost of Goods Sold (COGS) Analysis

```python
total_COGS_for_the_entire_dataset = dataset['Cogs'].sum()
total_COGS_for_the_entire_dataset
```

COGS compare to the target COGS per state

```python
COGS_compare_to_target_cogs = pd.DataFrame()
COGS_compare_to_target_cogs['Cogs'] = dataset['Cogs']
COGS_compare_to_target_cogs['Target_cogs'] = dataset['Target_cogs']
COGS_compare_to_target_cogs['State'] = dataset['State']
COGS_compare_to_target_cogs = COGS_compare_to_target_cogs.groupby(['State'])[['Cogs', 'Target_cogs']].sum().reset_index()
COGS_compare_to_target_cogs
```

3- Is there a correlation between COGS and profit?

```python
COGS_profet_correlation = pd.DataFrame()
COGS_profet_correlation['State'] = dataset['State']
COGS_profet_correlation['Cogs'] = dataset['Cogs']
COGS_profet_correlation['Profit'] = dataset['Profit']
COGS_profet_correlation = COGS_profet_correlation.groupby(['State'])[['Cogs', 'Profit']].sum().reset_index()
COGS_profet_correlation 
```

Marketing and Expenses Analysis

1- How much is spent on marketing in total?

```python
total_marketing_expenses = dataset['Marketing'].sum()
total_marketing_expenses
```

2- Is there a correlation between marketing expenses and sales or profit?

```python
marketing_sales_profit_correlation = pd.DataFrame()
marketing_sales_profit_correlation['Marketing'] = dataset['Marketing']
marketing_sales_profit_correlation['Sales'] = dataset['Sales']
marketing_sales_profit_correlation['State'] = dataset['State']
marketing_sales_profit_correlation['Profit'] = dataset['Profit']
marketing_sales_profit_correlation = marketing_sales_profit_correlation.groupby(['State'])[['Marketing', 'Sales', 'Profit']].sum().reset_index()
marketing_sales_profit_correlation
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

bar1 = ax.barh(np.arange(len(marketing_sales_profit_correlation)) - 0.2, marketing_sales_profit_correlation['Marketing'], 0.2, label='Marketing')
bar2 = ax.barh(np.arange(len(marketing_sales_profit_correlation)), marketing_sales_profit_correlation['Sales'], 0.2, label='Sales')
bar3 = ax.barh(np.arange(len(marketing_sales_profit_correlation)) + 0.2, marketing_sales_profit_correlation['Profit'], 0.2, label='Profit')

ax.set_ylabel('State')
ax.set_xlabel('Marketing')
ax.set_title('Marketing - Sales - Profit Correlation?')
ax.set_yticks(np.arange(len(marketing_sales_profit_correlation)))
ax.set_yticklabels(marketing_sales_profit_correlation['State'])
ax.invert_yaxis() 
ax.legend()

for bar in [bar1, bar2, bar3]:
    for value in bar:
        width = value.get_width()
        ax.annotate('{}'.format(width),
                    xy=(width, value.get_y() + value.get_height() / 2),
                    xytext=(3, 0),  # 3 points horizontal offset
                    textcoords="offset points",
                    ha='left', va='center')
```
